```python
from google_play_scraper import app, Sort, reviews, reviews_all
import pandas as pd
import numpy as np
```

```python
scrap = reviews_all(
  'com.openai.chatgpt', # App ID
  lang='id', # Language
  country='id', # Country
  sort=Sort.MOST_RELEVANT, # Sort order
  count=10000, # Number of reviews to fetch
  filter_score_with=None # Filter by score (1-5)
)
```

## Cek Data Scrapping

```python
data = pd.DataFrame(np.array(scrap),columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
data
```

## Sortir Kolom yang Dibutuhkan

```python
data_new = data[['userName', 'content', 'score', 'thumbsUpCount', 'at']]
data_new.head()
```
